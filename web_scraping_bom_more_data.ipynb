```python
from bs4 import BeautifulSoup
import requests
import time
```

```python
url_G = 'https://www.boxofficemojo.com/chart/mpaa_title_lifetime_gross/?by_mpaa=G'
url_PG = 'https://www.boxofficemojo.com/chart/mpaa_title_lifetime_gross/?by_mpaa=PG'
url_PG_13 = 'https://www.boxofficemojo.com/chart/mpaa_title_lifetime_gross/?by_mpaa=PG-13'
url_R = 'https://www.boxofficemojo.com/chart/mpaa_title_lifetime_gross/?by_mpaa=R'
```

```python
url_list = [url_G, url_PG, url_PG_13, url_R]
count = 0
movies = []
imdb_ids = []
```

```python
# main function, returns all relevant information we are looking for

def soupy(url, count, movies, imdb_ids):
    time.sleep(5)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    movies.extend(movie_deets(soup, movies, count))
    imdb_ids.extend(imdb_tags(soup, imdb_ids))
    print(len(imdb_ids))
    back = next_url(soup)
    if back:
        front = url.split('/chart')[0]
        new_url = front + back
        print(new_url)
        return soupy(new_url, count, movies, imdb_ids)
    elif count < 3:
        count +=1
        new_url = url_list[count]
        return soupy(new_url, count, movies, imdb_ids)
    else:
        return imdb_ids, movies
        
      
```

```python
def imdb_tags(soup, imdb_ids):
    movie_info = soup.findAll('tr', style='')
    imdb_id = []
    for i in range(1,len(movie_info)):
        for detail in movie_info[i].find('td', class_='a-text-left mojo-header-column mojo-truncate mojo-field-type-title'):
            imdb_id.append(detail.get('href').split('/')[2]) 
    return imdb_id
   
```

```python
# gets movie details in list format and returns to main function

def movie_deets(soup, movie_list, count):
    # navigates to where list of movies is
    movie_info = soup.findAll('tr', style='')
    movies = []
    rating = ['G', 'PG', 'PG-13', 'R']
    # loops through all movies in the list and adds info to master movie list
    for i in range(1,len(movie_info)):
        movie_deets = []
        for detail in movie_info[i].findAll('td'):
            movie_deets.append(detail.text)
        movie_deets.append(rating[count])
        movies.append(movie_deets)
    return movies
    
```

```python
# gets the next url to navigate to the next page of results

def next_url(soup):
    
    #navigates to where url is
    next_ur = soup.find('li', class_='a-last').find('a') 
    
    #if url is there will return the next url
    # once the end of results is reached will return False to main function in order to terminate it
    if next_ur:
        return next_ur.get('href')
    else:
        return False
        
```

```python
def get_columns(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    headers = soup.find('div', class_='a-section imdb-scroll-table-inner')
    heads = headers.find('table').findAll('th')
    
    columns = []
    for head in heads:
        columns.append(head.text.split('\n')[0])
    return columns
    
```
